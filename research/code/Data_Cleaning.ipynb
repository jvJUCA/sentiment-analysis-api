```python
import pandas as pd
```

```python
# df = pd.read_csv('/content/IMDB Dataset.csv',nrows=100)
df = pd.read_csv('/content/IMDB Dataset.csv')
print(len(df))
print(df.sample(5))
```

### Convert Labels 0/1

```python
def to_sentiment(rating):
  rating = str(rating)
  if rating == 'positive':
    return 0
  else:
    return 1

df['sentiment'] = df.sentiment.apply(to_sentiment)
print(df.sample(5))
```

### Save New Clean CSV

```python
# Assuming df is your DataFrame
df.to_csv('/content/IMDB Dataset_clean.csv', index=False)
```
